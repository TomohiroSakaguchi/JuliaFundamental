**プロット**

・プロットの基本

Pyplotを基本としているが，Juliaでは少し違ったやり方をしている．

ここでは```Plots.jl```の使い方をみていきましょう．

````julia
using Pkg
#Pkg.add("Plots")
using Plots
````

```Plots.jl```の利点の一つはシームレスなバックエンド変更が許されているという点です．

このノートブックでは```gr()```と```plotlyjs()```のバックエンドをやってみましょう．

````julia
x = -3:0.1:3
f(x) = x^2

y = f.(x)
````

GRバックエンドを読み込んでみましょう

````julia
gr()
````

````julia
plot(x, y, label="line")
scatter!(x, y, label="points")
````

ここでは```scatter!()```を使っているが，関数の後ろに```!```を挿入しないとプロットの上にプロットを重ねることができないからである．

```plotlyjs()```バックエンドを使ってみよう

````julia
#Pkg.add("PlotlyJS")
using PlotlyJS
plotlyjs()
````

````julia
plot(x, y, label="line")
scatter!(x, y, label="points")
````

```plotlyjs()```バックエンドを用いるとこのように直感的なグラフのプロットが可能である．

**少し凝ってみませんか？**

タイトルやラベルを構文に追加すると良くなりますよ．

ここでは地球平均気温と海賊の数の関係性をみていきましょう．(1860~2000)

````julia
globaltemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpirates = [45000, 20000, 15000, 5000, 400, 17]
````

````julia
#はじめにデータをプロットしましょう
plot(numpirates, globaltemperatures, legend=false)
scatter!(numpirates, globaltemperatures, legend=false)

#時系列にあるようにx軸の向きを反転させます
xflip!()

xlabel!("Number of pirates [Approximate]")
ylabel!("Global temperature (C)")
title!("Influence of pirates population on global warming")
````

subplotを生成したい時には個々でプロットをした後，layoutで明確にして，もう一回```plot()```をする必要があります．

````julia
p1 = plot(x,x)
p2 = plot(x,x.^2)
p3 = plot(x,x.^3)
p4 = plot(x,x.^4)
plot(p1, p2, p3, p4,layout=(2,2), legend=false)
````
